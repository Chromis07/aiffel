# 크롤러 만들기

```python
import urllib

def download(url):
    return urllib.request.urlopen(url)

# urlopen()함수를 통해 해당 서버에서 요청받은 ("HTTPResponse")객체를 반환
```

```python
from urllib.error import URLError, HTTPError, ContentTooShortError

def download(url):
    try:
        html = urllib.request.urlopen(url).read()
    except (URLError, HTTPError, ContentTooShortError) as e:
        
        # 요즘은 https도 잘 열림
        print('Download error', e.reason)
        html = None
    return html

download('https://www.google.com')
```

```python
def download2(url):
    try:
        response = requests.get(url)
        html = response.text
    except requests.ConnectionError:
        print('Connection error')
        html = None
    return html

download2('https://www.google.com')
```

# 환율 크롤링

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

#URL 가져오기
url = 'https://m.stock.naver.com/marketindex/index.nhn'
response = requests.get(url)

#Soup 객체 생성
soup = BeautifulSoup(response.text, 'html.parser')

# 국가 추출
country = []
stock_item = soup.findAll('span', {'class' : 'stock_item'})
for s in stock_item:
    country.append(s.text)

print(country)
```

```python
country.remove('달러인덱스')
```

```python
# 원하는 데이터 추출 - 환율정보
price = []
price_0 = soup.findAll('span', {"class":"stock_price"})
for p in price_0:
    price.append(p.text)

print(price)
```

```python
#데이터 정렬
data = {}
for i in range(len(country)):
    data[country[i]] = price[i+7]

pd.Series(data)
```
